```python
# importing libraries
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
from scipy.stats import sem
from statistics import mean, stdev
import scipy.stats as stats
```

```python
test1 = pd.read_excel (r'./lab5data_4jupyter.xlsx', index_col=0) 
test1
```

```python
vis1 = test1.visual1
aud1 = test1.auditory1
```

```python
VIS1_MEAN = mean(vis1)
VIS1_SEM = sem(vis1)
AUD1_MEAN = mean(aud1)
AUD1_SEM = sem(aud1)
```

```python
# build a bar plot
plt.bar(['Person 1 Visual', 'Person 1 Auditory'],[VIS1_MEAN, AUD1_MEAN], yerr=[VIS1_SEM, AUD1_SEM], capsize=10)
plt.ylabel('Speed (msec)')
plt.xlabel("Reflex Type")
plt.title('Reflex Speed, Visual vs. Auditory')
```

```python
#vis1 and aud1 refer to your columns that you are comparing (that you defined above)

res = stats.ttest_ind(vis1, aud1, 
                      equal_var=True)

display(res)
```

### p value of 0.0000179, which is significant! (anything less than 0.05 is significant)
