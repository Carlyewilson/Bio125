```python
# importing the required module 
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
#Optional for disabling warnings 
import sys
if not sys.warnoptions:
    import warnings
    warnings.simplefilter("ignore")
```

```python
df = pd.read_excel (r'./Fall 23 Physio Class Entire (1).xlsx', index_col=0) 
df.describe()
```

```python
df['Exam_percent'] = df.Exam1/38
df.describe()
```

```python
df['Exam_percent_curved'] = df.Exam1/38 + 0.044090
df.describe()
```

```python
%matplotlib inline
plt.rcParams.update({'figure.figsize':(7,5), 'figure.dpi':100})

# Plot Histogram on x
x1 = df.Exam_percent*100
plt.hist(x1, bins = 25, edgecolor='k')
plt.gca().set(title='Frequency Histogram - Original Score', ylabel='Frequency');
plt.xlim(0,100)

plt.axvline(x1.mean(), color='k', linestyle='dashed', linewidth=1)
min_ylim, max_ylim = plt.ylim()
plt.text(x1.mean()*1.1, max_ylim*0.9, 'Mean: {:.2f}'.format(x1.mean()))
```

```python
%matplotlib inline
plt.rcParams.update({'figure.figsize':(7,5), 'figure.dpi':100})

# Plot Histogram on x
x2 = df.Exam_percent_curved*100
plt.hist(x2, bins = 25, edgecolor='k')
plt.gca().set(title='Frequency Histogram - Curved Score', ylabel='Frequency');
plt.xlim(0,102)

plt.axvline(x2.mean(), color='k', linestyle='dashed', linewidth=1)
min_ylim, max_ylim = plt.ylim()
plt.text(x2.mean()*1.1, max_ylim*0.9, 'Mean: {:.2f}'.format(x2.mean()))
```

```python
kwargs = dict(hist_kws={'alpha':.4}, kde_kws={'linewidth':3})

plt.figure(figsize=(10,7), dpi= 300)
sns.distplot(x1, bins = 25, color="dodgerblue", label="Original", **kwargs)
plt.axvline(x1.mean(), color='dodgerblue', linestyle='dashed', linewidth=1)
min_ylim, max_ylim = plt.ylim()
plt.text(x1.mean()*.75, max_ylim*0.9, 'Mean: {:.2f}'.format(x1.mean()), color="dodgerblue")

sns.distplot(x2, bins = 25, color="orange", label="Curved", **kwargs)
plt.axvline(x2.mean(), color='orange', linestyle='dashed', linewidth=1)
min_ylim, max_ylim = plt.ylim()
plt.text(x2.mean()*1.10, max_ylim*0.9, 'Mean: {:.2f}'.format(x2.mean()), color="orange")
plt.xlim(20,110)
plt.legend();
plt.gca().set(title='Fall 2022 Human Anatomy - Exam 1 Scores - Before and After Curve', ylabel='Density', xlabel='Score')
```
