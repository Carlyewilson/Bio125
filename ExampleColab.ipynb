```python
import matplotlib.pyplot as plt
```

```python
VIS1_MEAN = 300.6
AUD1_MEAN = 574.7
VIS1_STD = 38.45
AUD1_STD = 145.07
```

```python
# build a bar plot
plt.bar(['Person 1 Visual', 'Person 1 Auditory'],[VIS1_MEAN, AUD1_MEAN], yerr=[VIS1_STD, AUD1_STD], capsize=10)
plt.ylabel('Speed (msec)')
plt.xlabel("Reflex Type")
plt.title('Reflex Speed, Visual vs. Auditory')
```

```python
VIS1_MEAN = 300.6
VIS2_MEAN = 574.7
VIS1_STD = 38.45
VIS2_STD = 145.07
```

```python
# build a bar plot
plt.bar(['Person 1 Visual', 'Person 2 Visual'],[VIS1_MEAN, VIS2_MEAN], yerr=[VIS1_STD, VIS2_STD], capsize=10)
plt.ylabel('Speed (msec)')
plt.xlabel("Person")
plt.title('Reflex Speed, Visual 1 vs 2')
```
